Sorting

```python
import pandas as pd
df1={"Name":["kamal","shruthi","Kumar"],
     "Marks":[80,95,89],
     "Age":[21,20,19]}
df1=pd.DataFrame(df1)
df1
```

```python
df1.sort_values(by=["Marks","Age"],inplace=True,ascending=False)
df1
```

```python
df1.sort_values(by="Name",key=lambda x: x.str.lower())
df1
```

Pivot table

```python
df = pd.DataFrame({'Product': ['Carrots', 'Broccoli', 'Banana', 'Banana',
                               'Beans', 'Orange', 'Broccoli', 'Banana'],
                   'Category': ['Vegetable', 'Vegetable', 'Fruit', 'Fruit',
                                'Vegetable', 'Fruit', 'Vegetable', 'Fruit'],
                   'Quantity': [8, 5, 3, 4, 5, 9, 11, 8],
                   'Amount': [270, 239, 617, 384, 626, 610, 62, 90]})
df
```

```python
pivot_df=pd.pivot_table(df,index=["Product","Category"],values=["Quantity","Amount"],aggfunc=["sum","mean","min",],margins=True)
pivot_df
```
