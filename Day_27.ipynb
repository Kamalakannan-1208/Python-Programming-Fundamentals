```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/bios.csv")
df.head()
```

```python
#groupby single col
group_by_borncity=df.groupby(by=["born_city"])
group_by_borncity
```

```python
#each grp first val
group_by_borncity.first()
```

```python
# grouped key
group_by_borncity.get_group("London").head()
```

```python
#grp by with agg
group_by_borncity.get_group("London")["height_cm"].mean()
```

```python
#multi col grpby
group_by_borncity_gender=df.groupby(by=["born_city","born_region"])
group_by_borncity_gender.first()
```

```python
# mul grp agg
group_by_borncity_gender.get_group(("London","England"))["weight_kg"].sum()
```
