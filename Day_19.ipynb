```python
import pandas as pd
dict_val={"name":["kamal","shruthi","Kumar","cavin","prakal"],
         "Age":[21,20,19,22,22],
         "dept":["cs","medico","mech","cs","cs"]}

df=pd.DataFrame(dict_val)
```

```python
df
```

Indexing and selecting data

```python
#[] operator
df["name"]
```

```python
df[["name","Age"]]
```

```python
#loc
df.loc[1,:]
```

```python
#iloc
df.iloc[[0,1],:]
```

Other useful methods

```python
# head
df.head()
```

```python
#tail
df.tail()
```

```python
# at and iat
df.at[1,"name"]
```

```python
#xs method
val=df.xs(0)
print(val)
```

```python
# get
# single
val=df.get("name")
print(val)
# multi
val=df.get(["name","Age"])
print(val)
```

```python
# isin
# single
val=df["Age"].isin([22])
print(val)
```

```python
print(df[val])
```

```python
#multi
val1=df["dept"].isin(["cs","mech"])
print(df[val&val1])
```

```python
# where
where_out=df.where(df["Age"]>20)
print(where_out)
```

```python
# mask
where_out=df.mask(df["Age"]>20)
print(where_out)
```

```python
# insert
new_col=tuple(["male","female","male","male","male"])
df.insert(1,value=new_col,column="gender")# if same order not needed keyword
print(df)
```

```python
# pop method
val=df.pop("dept")
print(val)
```
