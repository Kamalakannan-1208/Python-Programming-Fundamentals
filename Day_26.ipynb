```python
import pandas as pd
import numpy as np
df1=pd.DataFrame({"name":["kamal","kumar","kannan","suriya","cavin"],
                  "age":[22,22,"22",23,"23"]})
```

```python
df1.info()
```

```python
for i in df1.columns:
  print(f"{i} : {pd.api.types.infer_dtype(df1[i])}")
```
