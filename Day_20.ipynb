concat

```python
import pandas as pd
df1= {"Name":["kamal","shruthi"],
      "dept":["cs","Medical"]}
df2={"Name":["vijay","cavin"],
      "dept":["electrical","cs"]}
```

```python
df1=pd.DataFrame(df1)
df2=pd.DataFrame(df2)
concat=pd.concat([df1,df2],axis=0)
```

```python
# concat axis index as as
concat
```

```python
# resetting index
concat.reset_index(drop=True)
```

```python
# concat ignore index
concat1=pd.concat([df1,df2],axis=0,ignore_index=True)
concat1
```

```python
# concat by col
concat_column=pd.concat([df1,df2],axis=1)
concat_column
```

```python
# concat inner joins
# if u try axis=0 not working union
concat_inner=pd.concat([df1,concat1],axis=1,join="inner",ignore_index=True)
concat_inner
```

```python
# after concat also retain info then use
key_df=pd.concat([df1,df2],keys=["x","y"])
key_df
```

```python
# concat df and series
series=pd.Series(["A","B"])
concat_series=pd.concat([df1,series],axis=1)
concat_series
```

Merge(based on index or col)

```python
df1=pd.DataFrame({"Name":["kamal","shruthi","vijay","cavin"],
      "dept":["cs","Medical","electrical","cs"]})
df2=pd.DataFrame({"Name":["kamal","shruthi","cavin","renu"],
      "strong":["data","surgery","bug","singing"]})
```

```python
merge_name=pd.merge(df1,df2,on="Name")
merge_name
```

```python
merge_name_join=pd.merge(df1,df2,on="Name",how="outer")
merge_name_join
```

```python
merge_left =pd.merge(df1,df2,on="Name",how="left")
merge_left
```

```python
merge_right =pd.merge(df1,df2,on="Name",how="right")# arg left_on, right_on, on=[name,dept] multi key match
merge_right
```

Joining dataframe

```python
#joins
df1_join=df1.set_index("Name")
df2_join=df2.set_index("Name")
df1_join
```

```python
df2_join
```

```python
df1_join.join(df2_join,how="inner")
```
