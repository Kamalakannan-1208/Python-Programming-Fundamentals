Missing values


```python
import pandas as pd
import numpy as np
#df creation
df=pd.DataFrame({"name":["kamal","kuamar",None],
                 "age": [21,19,np.nan]})
df
```

```python
# checkign missing values
df.isnull()
```

```python
#checking not missing values
df.notnull()
```

```python
# count misssing values
df.isnull().sum()
```

```python
#retrieve or filter missing data rows
df[df.isnull().any(axis=1)]
```

```python
df[df.notnull().any(axis=1)]
```

```python
df["name"].fillna(method="ffill")
```

```python
df["age"].fillna(method="bfill")
```

```python
# replace
df.replace(to_replace=np.nan,value=0)
```

```python
# interplate
df.interpolate(method="linear",axis=0)
```
