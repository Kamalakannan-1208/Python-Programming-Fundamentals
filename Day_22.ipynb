```python
bios_dff=pd.read_csv("https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/data/bios.csv",sep=",",header=0,usecols=["athlete_id","name","born_city"],index_col="athlete_id",nrows=10, skiprows=[4,5])
bios_dff
```
