drop na

```python
import pandas as pd
import numpy as np
#df creation
df=pd.DataFrame({"name":["kamal",None,"kumar"],
                 "age": [21,19,np.nan]})
```

```python
# row based
df.dropna(axis=0,thresh=2)
```

```python
# col drop na
df.dropna(axis=1)
```

```python
df.dropna(how="all")
```

drop duplicates

```python
#
df=pd.DataFrame({"name":["kamal","kamal","kumar","kamal"],
                 "age": [21,21,np.nan,21]})
df
```

```python
df.drop_duplicates(keep="first")
```

```python
df.drop_duplicates(keep="last")
```

```python
df.drop_duplicates(subset="name",keep=False)
```
